# 2. DATA Transfer

```python
#Initial
n = 153
m = 200000
a = [[0]*n for x in range(m)]
#print(a)
# index 0: starttime 1: chamber 2:sheetid
a[0][0] =(data['STARTTIME'][0])
a[0][1] = data['SHEETID'][0]
a[0][2] = data['CHAMBER'][0]
```

```python
# first data insert to list 
starttime = a[0][0]
j = 3
i = 0
index = 0
bstore = 0 # check if get 150 point data

#if i <= data.shape[0]:
while i < data.shape[0]:
        # data['STARTTIME'][i] : STARTTIME
        if bstore == 0:
            for j  in range(150):
                if starttime == data['STARTTIME'][i]:
                    a[index][j + 3]=  data['TORQUE'][i]
                else :
                    a[index][j + 3] = 0
                i = i + 1 
            bstore = 1
        if starttime != data['STARTTIME'][i]:
                bstore = 0
                starttime =   data['STARTTIME'][i]
                a[index][0] = starttime
                a[index][1] = data['SHEETID'][i]
                a[index][2] = data['CHAMBER'][i]
                index = index + 1
        i = i + 1
            
```
